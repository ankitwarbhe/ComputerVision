### Import Libraries

```python
import  cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Create Classifiers

```python
face_classifier=cv2.CascadeClassifier("Haarcascades/haarcascade_frontalface_default.xml")
eyes_classifier=cv2.CascadeClassifier("Haarcascades/haarcascade_eye.xml")
```

### Open Image

```python
img=cv2.imread("Images/eye_face.jpg")
```

### Fix Image

```python
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### Face Classifier

```python
face=face_classifier.detectMultiScale(fix_img,1.3,5)
```

### If No Face Detected

```python
if face is ():
    print("No Faces Found")
```

### It's a Kind of Magic

```python
def detect_faces_eyes(fix_img):
    
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w,y+h),
                     (255,0,0),
                     7)
      
        eyes_rects =eyes_classifier.detectMultiScale(fix_img)
        for(ix,iy,iw,ih) in eyes_rects:
            cv2.rectangle(fix_img,
                     (x,y),
                     (x+w,y+h),
                     (0,0,255),
                     5)
        return fix_img
```

```python
result=detect_faces_eyes(fix_img)
plt.imshow(result)
```
